## Importing libraries

```python
import pandas as pd
```

## Task 1: Explore Individual Datasets

```python
cust = pd.read_csv("Global Superstore 3NF/Customers.csv")
order = pd.read_csv("Global Superstore 3NF/Orders.csv")
prod = pd.read_csv("Global Superstore 3NF/Products.csv")
```

### Key Columns
The key columns that link the data sets are Customer ID, Order ID and Product ID; the data can be integrated with these columns.

```python
cust.head()
```

```python
order.head()
```

```python
prod.head()
```

### Data Quality Issues
The ID column names are consistent across the data sets. The ID data types are also consistent - all objects. 

There are fewer unique IDs than there are total rows for every dataset. Since these IDs are not actually valid (unique) keys, the data sets need to be cut down so the IDs are unique before joining on them.

## Task 2: Data Preparation

### Deduplicating IDs

```python
cust_dedup = cust.drop_duplicates(subset=['Customer ID'])
```

```python
cust_dedup.describe()
```

```python
order_dedup = order.drop_duplicates(subset=['Order ID'])
```

```python
order_dedup.describe(include='object')
```

```python
prod_dedup = prod.drop_duplicates(subset=['Product ID'])
```

```python
prod_dedup.describe(include='object')
```

## Task 3: Data Integration

### Joining (error)
There is an error when joining data sets that have a mix of int and object data types. There is legitimately numerical data in the orders data set, so concat or merge needs to be used instead.

```python
order_prod_dup_join = order.join(prod, on='Product ID', how='left')
```

### Merge Orders with Products

First, merging on the data without deduplication. See how the total number of Order IDs actually increases because of the duplicate order-product ID pairs.

```python
order_prod_dup_merge = pd.merge(order, prod, on='Product ID')
```

```python
order_prod_dup_merge.describe(include='object')
```

Now with the deduplicated data:

```python
order_prod_merge = pd.merge(order_dedup, prod_dedup, on='Product ID')
```

```python
order_prod_merge.describe(include='object')
```

### Merge vs Concat of Order-Product Table with the Customers table

Note that the concat join parameter specifies handling of *indexes* on the rows axis (since the data is concat on axis 1 - columns). Therefore, when specifying an inner join for concat, the data is limited to the set with the smallest index (Customers). And on the other hand, when performing a concat with an outer join, the customer data is not paired with all the corresponding order rows - the count of Customer ID is much lower than Order ID.

```python
order_prod_cust_merge = pd.merge(order_prod_merge, cust_dedup, how='inner', on='Customer ID')
```

```python
order_prod_cust_merge.describe(include='object')
```

```python
order_prod_cust_concat_inner = pd.concat([order_prod_merge, cust_dedup], axis=1, join='inner')
```

```python
order_prod_cust_concat_inner.describe(include='object')
```

```python
order_prod_cust_concat_outer = pd.concat([order_prod_merge, cust_dedup], axis=1, join='outer')
```

```python
order_prod_cust_concat_outer.describe(include='object')
```

## Task 4: Post-Integration Review

Only Postal Code has missing values, so the integration was performed correctly. Postal codes apparently were not necessary for every order. As long as analysis doesn't focus on postal codes, then the data is good for analysis. 

```python
order_prod_cust_merge.isna().sum()
```

The inner and outer joined concat data sets are missing much information, though the outer joined concat data has completely missing values for much of the customer data.

```python
order_prod_cust_concat_outer.isna().sum()
```
